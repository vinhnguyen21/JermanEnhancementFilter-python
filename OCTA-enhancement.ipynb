```python
from PIL import Image
import numpy as np
import cv2
import matplotlib.pyplot as plt
from vesselness2d import *
```

```python
img_dir = 'images/test.tif'
```

```python
#reading image
image = Image.open(img_dir).convert("RGB")
image = np.array(image)
plt.figure(figsize=(10,10))
plt.imshow(image, cmap='gray')

#convert forgeground to background and vice-versa
image = 255-image

image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
thr = np.percentile(image[(image > 0)], 1)*0.9
image[(image <= thr)] = thr
image = image - np.min(image)
image = image / np.max(image)
```

```python
# enhance vesselness with Jerman's  2D Hessian based tubular (vessel/vesselness) and spherical (blob/blobness) enhancement filters.

#hyper-parameter
sigma=[0.5,1, 1.5, 2, 2.5]
spacing = [1, 1]
tau = 2

output = vesselness2d(image, sigma, spacing, tau)
output = output.vesselness2d()
```

```python
plt.figure(figsize=(10,10))
plt.imshow(output, cmap='gray')
```
